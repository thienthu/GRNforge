# Construction of a proper eGRN class
Would it not be better to make different generic constructors, and add raw data to the eGRN class?
In that way, I could decice for a different threshold on some variable, and the class re-constructs
my filter-specific GRN. That would make it quite flexible in terms of testing differetn settings. Each
input table will (for now) be represented as pandas dataframe, and therefore allow for arbitrary filtering later on. Especially now, we want a 'base-GRN' class that supportst the efficient sub-setting of regression tasks downstream  
- First input:    TF-motif mapping table (pre-computable)
- Second input:   Full peak space (+ optional motif footprints per peak)
- Third input:    Full TFBS space annoatead with their respective peak 
- Fourth input:   A distance - relationship mapping between TGs and peaks (pre-computable)
  
After that, we need one function per eGRN layer to compile it. For developing reasons, I wil add the class here now t

```python
# Class to further implement eGRN structures that can be used by other methods

class EnhancerGRN:
    """
    Core enhancer GRN dictionary structure
    Lets make it as simple as possible, and reverse each dictioary order
    for maximal lookup speed (given the slowness of python of course)
    """
    def __init__(self):
        
        # pandas dataframe for the full data
        TF_motif_data     = None
        Full_peak_data    = None
        Full_TFBS_data    = None
        Full_TG_peak_data = None
        
        
        # TF - Motif layer and inverse
        self.tf_tfbs_map = {}  # connects all TFs to a TFBS (motif) 
        self.tfbs_tf_map = {}  # connects a TFBS (motif) to all its motifs 
        # TFBS - RE layer and inverse
        self.tfbs_re_map = {}  # map each tfbs to all its REs
        self.re_tfbs_map = {}  # map each RE to all its TFBS
        # TF - RE layer and inverse
        self.tf_re_map   = {}  # map each tfbs to all its REs
        self.re_tf_map   = {}  # map each RE to all its TFBS
        # TG - RE layer and inverse
        self.tg_re_map   = {}  # map each tfbs to all its REs
        self.re_tg_map   = {}  # map each RE to all its TFBS
        
        
```

```python
# start the annotation with TOBIAS footprints
import pyBigWig
bw = pyBigWig.open("/scratch/gent/vo/000/gvo00027/projects/CBIGR/24JLR_GRN/results/human_PBMC_10x_10k/pbmc_granulocyte_sorted_10k_atac_peaks_footprints.bw")
```

```python
scores_mean = []
scores_sum  = []
scores_max  = []

# score all peaks
for chrom, start, end in zip(peaks['chrom'], peaks['start'], peaks['end']):
    try:
        values = bw.values(chrom, start, end)
        scores_mean.append(np.mean(values))
        scores_sum.append(np.sum(values))
        scores_max.append(np.max(values))
    except:
        print('skip entry', chrom, start, end)
        scores_mean.append(0)
        scores_sum.append(0)
        scores_max.append(0)
        
    
```

```python
# add to file
peaks['ftp_scores_mean'] = np.nan_to_num(scores_mean)
peaks['ftp_scores_sum']  = np.nan_to_num(scores_sum)
peaks['ftp_scores_max']  = np.nan_to_num(scores_max)
```

# 3. TSS peak annotation

```python
# load HG38 annotation file to receive IDs
infile = open('/scratch/gent/vo/000/gvo00027/projects/CBIGR/24JLR_GRN/data/universal_files/gencode.v38.annotation.gff3','r')
new_lines = []

for line in infile.read().split('\n'):
    if len(line) == 0 or line [0] == '#':
        continue
    
    else:
        line_info = []
        info      = line.split('\t')
        line_info.append(info[0])
        line_info.append(info[3])
        line_info.append(info[4])
        line_info.append(info[2])
        line_info.append(info[6])
        tmp = info[8].split(';')
        id_1      = '.'
        gene_name = '.'
        gene_type = '.'
        hgnc_id   = '.'
        for i in tmp:
            if i.startswith('ID='):
                id_1      = i.split('=')[1]
            if i.startswith('gene_type'):
                gene_type = i.split('=')[1]
            if i.startswith('gene_name'):
                gene_name = i.split('=')[1]     
            if i.startswith('hgnc_id'):
                hgnc_id = i.split('=')[1]     
        line_info.append(id_1)
        line_info.append(gene_name)
        line_info.append(gene_type)
        line_info.append(hgnc_id)    
    new_lines.append(line_info)
    
genome_annotation = pd.DataFrame(new_lines)
```

```python
# now intersect them to get the annotations

# Create BedTool objects for the two BED files
bedfile1 = pybedtools.BedTool('/scratch/gent/vo/000/gvo00027/projects/CBIGR/24JLR_GRN/results/human_PBMC_10x_10k/peaks.bed')  # Gene regions with thresholds
bedfile2 = pybedtools.BedTool('/scratch/gent/vo/000/gvo00027/projects/CBIGR/24JLR_GRN/results/human_PBMC_10x_10k/tss.bed')  # Peak locations

# Intersect peaks with gene regions, including annotations from bedfile1
intersected = bedfile1.intersect(bedfile2, wa=True, wb=True)

# Save the annotated results or process further
intersected.saveas('/scratch/gent/vo/000/gvo00027/projects/CBIGR/24JLR_GRN/results/human_PBMC_10x_10k/TSS_annotated_peaks.bed')

```
